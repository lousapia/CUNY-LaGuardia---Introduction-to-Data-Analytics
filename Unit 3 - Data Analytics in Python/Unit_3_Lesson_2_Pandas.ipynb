```python
# Work with missing values (A and B are column names)
df = {'A':[1,2,np.nan], 'B':[3,np.nan,np.nan]}
df
df = pd.DataFrame(df)
df
```

```python
# check for missing values for each column
df.isnull().sum()
```

```python
# returns the number of all missing values
print(df.isnull().sum().sum())
```

```python
# Removing missing values
# Returns only rows with no missing values
df.dropna() 
```

```python
# Drops NAs only from columns that contain NA values.  If a column doesn’t have any missing values, it wont be dropped.
df.dropna(axis=1) 
```

```python
# Sets a threshold – NAs will only be dropped if that threshold is met.
df.dropna(thresh=1) 
```

```python
# Fill the missing value with the one you specify. 
df.fillna(value='empty') 
```
