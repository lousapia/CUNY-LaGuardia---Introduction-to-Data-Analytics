```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/realpython/python-data-cleaning/master/Datasets/BL-Flickr-Images-Book.csv')
df.head()
```

```python
# You can create a list of columns that you’d like to drop.
# Then, pass that list into the drop() function
# Inplace=True tells Python to remove those columns from our current variable, df
# Axis=1 tells Python to look on the column axis
drop_columns = ['Edition Statement','Corporate Author','Corporate Contributors','Former owner','Engraver']
df.drop(drop_columns, inplace = True, axis = 1)
```

```python
df.head()
```

```python
# Check to see if Identifier is unique
df['Identifier'].is_unique
```

```python
# Set the index to the identifier
df = df.set_index('Identifier')
df.head()
```

```python
df1 = df['Date of Publication']
df1
```

```python
# r'^(\d{4})’  
# looking for:
#     ^       a string that starts with...
#      \d    a digit (0-9)...
#     {4}   that repeats four times...

#expand bool, default True
# If True, return DataFrame with one column per capture group. 
# If False, return a Series/Index if there is one capture group or 
# DataFrame if there are multiple capture groups.

df1 = df['Date of Publication'].str.extract(r'^(\d{4})', expand = False  )
df1
```

```python
# datatype shows it is of object datatype
df['Date of Publication'].dtype
```

```python
# Convert to numeric
df['Date of Publication'] = pd.to_numeric(df1)
df['Date of Publication'].dtype
```

```python
# Non-stardard missing values 
# -, --, na, n/a
# This will replace the non-standard missing values with standard missing value (NaN)
missing_values = ['-', '--', 'na', 'n/a']
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/property.csv', na_values = missing_values)
df.head()
```

```python
# look at the data and see the odd data for NUM_BATH and OWN_OCCUPIED
df
```

```python
# Coerce will change unexpected value to NaN:
df['NUM_BATH'] = pd.to_numeric(df['NUM_BATH'],errors='coerce')
df
```

```python
# Replace missing values with a number
df['ST_NUM'].fillna(125,inplace=True)
df
```

```python
# Replace certain values by location
df.loc[4,'ST_NUM'] = 125
df
```

```python
# replace using median
median = df['NUM_BEDROOMS'].median()
df['NUM_BEDROOMS'].fillna(median,inplace=True)
df
```

```python
# Drop a row by index number
df = df.drop(index=201)
df
```
